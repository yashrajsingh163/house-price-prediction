```python
import numpy as np
import pandas as pd
```

```python
#import data
data =  pd.read_csv(r"data.csv")

#print 5 records 
data.head()

#statistical analysis
data.describe()

#data infomation
# data.info()

#checking any null value 
data.isnull().sum()
```

# DATA Processing


```python
#drop unwanted columns and unnessessary columns 
data_modify = data.drop(['date','street','statezip','country'],axis=1)

#again check data threre is any object column 
data_modify.info()

#there is one categorial column avaliable is city so we need to check how many citys avaliabe in data set
data_modify['city'].unique()
```

```python
#handle city categorical variable 
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data_modify['city'] = le.fit_transform(data_modify['city'])
```

```python
data_modify.info()
```

```python
#divide data target and non target columns
X = data_modify.drop("price", axis=1)
Y = data_modify['price']
```

```python
#divide data into tranning and testing datasets
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)  #80 % of traning and 20% of testing 
```

```python
#creating the machine learning model
from sklearn.linear_model import LinearRegression
linearmodel=LinearRegression()
linearmodel.fit(X_train,Y_train) # training is taking place
```

```python
linearmodel.predict(X_test)
```

```python
#calculating the accuracy of the model
linearmodel.score(X,Y) # evaluation is taking place
```

```python
#save model
import pickle

with open('house_price2', 'wb') as files:
    pickle.dump(linearmodel, files)
```
